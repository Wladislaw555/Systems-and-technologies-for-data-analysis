```python
print("Задание по матрицам 2")

def check_dimensions(A, B):
    """
    Проверка размерностей матриц A и B
    """
    rows_A, cols_A = len(A), len(A[0])
    rows_B, cols_B = len(B), len(B[0])
    return rows_A == rows_B and cols_A == cols_B

def sum_matrices(A, B):
    """
    Сумма матриц A и B
    """
    return [[A[i][j] + B[i][j]  for j in range(len(A[0]))] for i in range(len(A))]

def calculate_and_print_sum(A, B, name):
    """
    Calculate and print the sum of matrices A and B
    """
    if check_dimensions(A, B):
        C = sum_matrices(A, B)
        print(f"\nСумма матриц {name}:")
        for row in C:
            print(row)
    else:
        print(f"\nСумма матриц {name} не может быть найдена, потому что они имеют разные размерности.")

# Определение матриц A и B
A = [[1, 2]]
B = [[3], [0]]

A1 = [[0, 1]]
B1 = [[0], [2], [3]]

A2 = [[1, -1], [0, 2]]
B2 = [[2, 3], [4, -3]]

A3 = [[1, 2, 0, 1], [-1, 3, -2, 4]]
B3 = [[2, 5, -1, 3], [3, 0, -2, 4]]

# Calculate and print sums
calculate_and_print_sum(A, B, "A и B")
calculate_and_print_sum(A1, B1, "A1 и B1")
calculate_and_print_sum(A2, B2, "A2 и B2")
calculate_and_print_sum(A3, B3, "A3 и B3")
```
